### Assumption  
This will be solved classification problem. Annual salary will be classified "High" if over median salary and "low" if under median salary. Target feature will be labeled 'high_sal', and high salary classified as 1 while low salary classified as 0.

```python
cutoff = finaldf['annual_mean_salary'].median()
finaldf['high_sal'] = finaldf['annual_mean_salary'].apply(lambda x : 1 if x>=cutoff else 0 if x<cutoff else np.nan)
finaldf = finaldf.drop(['annual_mean_salary','annual_salary_range'], axis=1)
finaldf
```

### Part 1: Analyzing Job Titles.  
#### Applying Count Vectorizer on the job titles.

```python
from sklearn.feature_extraction.text import CountVectorizer
cvz = CountVectorizer(stop_words='english',ngram_range=(1, 2))
cvz.fit(finaldf['job_title'])

titleWords = pd.DataFrame(cvz.transform(finaldf["job_title"]).todense(),
                       columns=cvz.get_feature_names())

titledf=pd.concat([titleWords,finaldf['high_sal']],axis=1)
titledf.head()
```

```python
#extracting the salary NaN values as a holdout set for prediction later.
title_hold = titledf.loc[titledf['high_sal'].isnull()]
titledf = titledf.drop(titledf.index[titledf['high_sal'].isnull()])
title_hold.head()
```

```python
#Checking the shape
titledf.shape
```

```python
#Making sure 'high_sal' is integer. Couldn't do that with NaN values earlier.
titledf['high_sal']=titledf['high_sal'].astype(int)
titledf.high_sal.dtype
```

```python
title_X = titledf.drop('high_sal',axis=1)
title_y = titledf['high_sal']
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(title_X, title_y, test_size=0.25, stratify=title_y, random_state=28)
```

#### Applying Random Forest Classifier

```python
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
```

```python
rf = RandomForestClassifier(random_state=28)  

scores = cross_val_score(rf, X_train, y_train, cv=10)
print("Cross-validated scores for train set:", scores)
print("Mean cross-validated score:", scores.mean())

#Fitting the model
rfmodel = rf.fit(X_train, y_train)

#Predictions/probabilities on the test dataset
y_pred = pd.DataFrame(rfmodel.predict(X_test))  
y_prob = pd.DataFrame(rfmodel.predict_proba(X_test))

title_rf_acc = accuracy_score(y_test, y_pred)  
title_rf_roc_auc = roc_auc_score(y_test, y_prob[1])
print("Accuracy for test set:",title_rf_acc)
print("ROC AUC Score for test set:",title_rf_roc_auc)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Feature Importance
pd.Series(rfmodel.feature_importances_, index=X_test.columns).sort_values(ascending=False).head(20)
#'lead' seems to be the most important word in job titles, followed by 'scientist' and 'senior'. Makes sense. 
```

```python
#Predicting 'high_sal' for the dataset without salary data. This part has no score for obvious reasons.
title_hold = title_hold.drop('high_sal',axis=1)
y_pred_title = pd.DataFrame(rfmodel.predict(title_hold))
nosalaryheader = df.loc[title_hold.index,['job_title','company']]
title_pred = pd.concat([nosalaryheader.reset_index(drop=True),y_pred_title], axis=1)
title_pred.columns = ['job_title','company','title_pred']
title_pred.head(10)
```

#### Insights  
Based on job title alone, this Random Forest Classifier is able to give a pretty good prediction of salary with almost 71% accuracy.  
The most important keywords in a job title that's highly related to whether salary is high or low are keywords that relate to seniority, such as 'lead' and 'senior'. 'scientist' is also rated pretty high, indicating that scientists in general earn higher than median salary compared to other positions like 'analyst' and 'engineer'.

### Part 2: Distinguishing Junior from Senior Roles.  
#### Feature Engineering  
We'll reuse the alldf2 dataset defined earlier. There's already a categorical variable for junior and senior roles, we'll use that to define our target variable, then drop those variables.

```python
seniordf = alldf2.copy()
```

```python
#Assigning job seniority level to Boolean masks.
s_mask = (seniordf['Senior Executive']+seniordf['Manager']+seniordf['Senior Management'])>0
j_mask = (seniordf['Fresh/entry level']+seniordf['Junior Executive']+seniordf['Executive'])>0
```

```python
seniordf['is_senior'] = [1 if i in seniordf.index[s_mask] else 0 if i in seniordf.index[j_mask] else np.nan for i in seniordf.index]
```

```python
#dropping the NaN values for roles not assigned to senior or junior status
seniordf = seniordf.drop(seniordf.index[seniordf['is_senior'].isnull()])
```

```python
#dropping the variables that already carry the same info
seniordf = seniordf.drop(['Senior Executive','Manager','Senior Management','Fresh/entry level','Junior Executive','Executive'], axis=1)
seniordf
```

```python
seniordf.groupby(['is_senior'])['Contract'].count().plot(kind='bar')
#Slight imbalance. Should not be a problem.
```

#### Applying Bernoulli Naive Bayes

```python
#Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.25, stratify=y2, random_state=28)
```

```python
nb = BernoulliNB()  

scores = cross_val_score(nb, X_train, y_train, cv=10)
print("Cross-validated scores for train set:", scores)
print("Mean cross-validated score:", scores.mean())

#Fitting the model
nbmodel = nb.fit(X_train, y_train)

#Predictions/probabilities on the test dataset
y_pred = pd.DataFrame(nbmodel.predict(X_test))  
y_prob = pd.DataFrame(nbmodel.predict_proba(X_test))

nb_acc = accuracy_score(y_test, y_pred)  
nb_roc_auc = roc_auc_score(y_test, y_prob[1])
print("Accuracy for test set:",nb_acc)
print("ROC AUC Score for test set:",nb_roc_auc)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
