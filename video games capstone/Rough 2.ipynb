```python
def random_dates(start, end, n=10):

    start_u = start.value//10**9
    end_u = end.value//10**9

    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit='s')
```

AI Singapore had an interesting one. Write a function to convert a number from 0 to 999,999 to its alpha equivalent (Punctuations included)

E.g. 1227 - One thousand, two hundred and twenty seven)

```python
def to_alpha(number):
    #Defining the dictionaries
    singleteen = {1:'one', 2:'two', 3:'three', 4:'four', 5:'five', 6:'six', 7:'seven', 8:'eight', 9:'nine', 10:'ten',\
              11:'eleven', 12:'twelve', 13:'thirteen', 14:'fourteen',15:'fifteen', 16:'sixteen', 17:'seventeen', 18:'eighteen',19:'nineteen'}

    tens = {20:'twenty',30:'thirty',40:'forty',50:'fifty',60:'sixty',70:'seventy',80:'eighty',90:'ninety'}

    #Initializing the output
    output = ''

    #Hundred thousandth place
    if len(str(number))>5:
        output = output+singleteen[int(number/100000)]+' hundred'
    else:
        pass

    #Thousandths place. Ten thousandths place is bundled here, only handled separately if the 2 digit number is more than 20.
    #Because 'eleven' to 'twenty' is annoying.
    if len(str(number))>=4:
        thousandth = int((number%100000)/1000)
        if thousandth > 0: 
            output = output+' and '
     
        if thousandth > 20:
            tenthousands = int(thousandth-thousandth%10)
            thousands = int(thousandth%10)
            output = output+tens[tenthousands]+' '+singleteen[thousands]+' thousand'
        elif thousandth <= 20 and thousandth >0:
            output = output+singleteen[thousandth]+' thousand'
        elif thousandth == 0:
            output = output+' thousand'
    else:
        pass

    #Hundredth place
    if len(str(number))>2:
        hundreds = int(number%1000)
            
        if hundreds > 0 and len(str(number))>3:
            if hundreds < 100 or hundreds%100 == 0:
                output = output+' and '
            else:
                output = output+', '
        elif hundreds > 0:
            output = output+singleteen[int(hundreds/100)]+' hundred'
    else:
        pass

    #Tens and ones bundled together, only handled separately if the 2 digit number is more than 20.
    #Because 'eleven' to 'twenty' is annoying.
    if len(str(number))>=1:
        if len(str(number))>2 and number%100 > 0: 
            output = output+' and '
        ten_ones = number%100
        if ten_ones > 20:
            ten = int(ten_ones-ten_ones%10)
            ones = int(ten_ones%10)
            output = output+tens[ten]+' '+singleteen[ones]
        elif ten_ones > 0:
            output = output+singleteen[ten_ones]
        elif number == 0:
            output = 'zero'
            
    if output[-1] != '.':
        output = output+'.'
    
    return output
```
