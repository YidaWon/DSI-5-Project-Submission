```python
def cleanit(df):
    #Drop all the locked sales data with 'Pro' value
    df.drop(df.index[df['Weekly_Sales']=='Pro'],inplace=True)
    
    #Drop all NaN values for game names
    df.drop(df.index[df['Name'].isnull()],inplace=True)
    
    #Change 'Week_Ending' to datetime format
    df['Week_Ending'] = pd.to_datetime(df['Week_Ending'])
    #Getting Year, Month and Day info from datetime
    df['Year'] = df['Week_Ending'].apply(lambda x : x.year)
    df['Month'] = df['Week_Ending'].apply(lambda x : x.month)
    df['Date'] = df['Week_Ending'].apply(lambda x : x.day)
    
    #Changing Sales numbers and Week No. to integers
    df['Weekly_Sales'] = df['Weekly_Sales'].apply(lambda x: int(x.replace(',','')))
    df['Total_Sales'] = df['Total_Sales'].fillna(str(0))
    df['Total_Sales'] = df['Total_Sales'].apply(lambda x: int(x.replace(',','')))
    df['Week_No.'] = df['Week_No.'].apply(lambda x: int(x))
    
    return df
```

```python
gclean['Name_Console'].nunique()
```

```python
#Pivoting Weekly Sales Numbers. Objective is to sum up the weekly sales numbers for Total Sales if current Total Sales numbers don't ally
gpiv = gclean.groupby(['Name_Console','Week_No.'])[['Weekly_Sales']].sum().reset_index()
gpiv = gpiv.pivot(index='Name_Console',columns='Week_No.',values='Weekly_Sales').reset_index()
gpiv.tail(10)
```

```python
#Pivoting for total sales. Since Total Sales is given per week, the numbers here should be cumulative
gpivtot = gclean.groupby(['Name_Console','Week_No.'])[['Total_Sales']].sum().reset_index()
gpivtot = gpivtot.pivot(index='Name_Console',columns='Week_No.',values='Total_Sales').reset_index()
gpivtot.tail(10)
```

```python
#Filling up null values for 1st Week (To be used later)
gpivtot[1].fillna(gpivtot[0], inplace=True)
gpivtot[1].fillna((gpivtot[3]-gpivtot[2]), inplace=True)
gpivtot[1].isnull().sum()
```

```python
#Appending the max value from total sales to gpiv (Max as it's cumulative)
gpiv['Total_Sales'] = gpivtot.max(axis=1)
gpiv
```

```python
#Checking for discrepancies 
gpiv.loc[gpiv['Total_Sales']<gpiv[[0,1,2,3,4,5,6]].sum(axis=1)]
```

```python
#Replacing the Total Sales in these entries with the sum
for i in gpiv.index[gpiv['Total_Sales']<gpiv[[0,1,2,3,4,5,6]].sum(axis=1)]:
    gpiv.loc[[i],'Total_Sales'] = gpiv[[0,1,2,3,4,5,6]].sum(axis=1)
gpiv.loc[gpiv['Total_Sales']<gpiv[[0,1,2,3,4,5,6]].sum(axis=1)]
```

```python
#Adding in the first week Total Sales data as an extra feature
gpiv['First_Week'] = gpivtot[1] 
gpiv
```
