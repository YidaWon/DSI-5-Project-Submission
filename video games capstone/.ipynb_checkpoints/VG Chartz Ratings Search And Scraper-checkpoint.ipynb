```python
#importing libraries for scraping
import pandas as pd
import urllib
from bs4 import BeautifulSoup
import os
from selenium import webdriver
import requests
import numpy as np
from time import sleep
from datetime import datetime
```

```python
gfinal = pd.read_csv('./gfinal.csv')
gfinal = gfinal.set_index('Name_Console')
```

```python
gfinal
```

```python
def getscore(title,console):
    title = urllib.parse.quote_plus(title)
    url = 'http://www.vgchartz.com/games/games.php?name='+title+'&keyword=&console='+console+'&region=All&developer=&publisher=&goty_year=&genre=&boxart=Both&banner=Both&ownership=Both&results=200&order=Sales&showtotalsales=0&showtotalsales=1&showpublisher=0&showpublisher=1&showvgchartzscore=0&shownasales=0&showdeveloper=0&showdeveloper=1&showcriticscore=0&showcriticscore=1&showpalsales=0&showreleasedate=0&showreleasedate=1&showuserscore=0&showuserscore=1&showjapansales=0&showlastupdate=0&showothersales=0'
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')
    
    searchresult = soup.find('tr',{'style':'background-image:url(../imgs/chartBar_large.gif); height:70px'})
    for attempts in range(3):
        try:
            criticscore = searchresult.findAll('td')[6].text.strip()
            userscore = searchresult.findAll('td')[7].text.strip()
            break
        except:
            sleep(1)
    else:
        criticscore = np.nan
        userscore = np.nan
        
    return {'Critic_Score':criticscore,'User_Score':userscore}
```

```python
#testing function
getscore("AKB1/48: Idol to Guam de Koishitara...",'PSP')
```

```python
searchterms = gfinal[['Name','Platform']]
```

```python
result = pd.DataFrame(columns = ['Name_Console','Critic_Score','User_Score'])
result
```

```python
searchterms.loc["A.C.E.: Another Century's Episode R (PS3)",'Name']
```

```python
name = searchterms.loc["A.C.E.: Another Century's Episode R (PS3)",'Name']
platform = searchterms.loc["A.C.E.: Another Century's Episode R (PS3)",'Platform']
getscore(name, platform)
```

```python
count = 0
skip = 0
for game in searchterms.index:
    name = searchterms.loc[game,'Name']
    platform = searchterms.loc[game,'Platform']
    for attempts in [1,2]:
        try:
            scores = getscore(name,platform)
            scores['Name_Console'] = game
            result = result.append(scores, ignore_index = True)
            count += 1
            print('Game No.',str(count)+',',game,'scores appended.',attempts,'tries.',scores['Critic_Score'],scores['User_Score'])
            break
        except:
            sleep(1)
    else:
        skip += 1
        print(game,'skipped.')
print(count, 'games appended.', skip,'games skipped.')
```
