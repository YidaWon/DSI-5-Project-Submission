```python
#importing libraries for scraping
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
from time import sleep
from datetime import datetime
```

```python
#We know from vgchartz that the weekly sales record goes back to 23rd Apr 1989. This corresponds to a page index of 32621.
#The index increment by 7 for every week
#Setting the range
JapanDateRange = list(range(32621,43346,7))
```

```python
def perweek(page):
    df = pd.DataFrame(columns = ['Name','Platform','Publisher','Genre','Weekly_Sales','Total_Sales','Week_No.','Week_Ending'])

    #page needs to be a string
    page = str(page)
    
    #Parsing into Beautiful Soup
    url = 'http://www.vgchartz.com/weekly/'+page+'/Japan/'
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')

#get week ending
    week_ending = soup.find('option',{'value': page}).text

#Getting the relevant table
    chartsoup = soup.find('div',{'id':'chart_body'})
    table = chartsoup.find('table',{'class':'chart'})

    for tr in table.findAll('tr')[1::2]:
        #Had to increment by 2 for this for loop as there's an extra "tr" element within each "tr" containing each game's stats.
        td = tr.findAll('td')
        weekly = td[4].text
        total = td[5].text
        week_no = td[6].text
        name = td[3].findAll('a')[0].text
        platform = td[3].findAll('a')[1].text
        publisher = td[3].findAll('a')[2].text
        genre = td[3].text.split(',')[-1].strip()
        
        df = df.append({'Name':name, 'Platform':platform, 'Publisher':publisher, 'Genre':genre, 'Weekly_Sales':weekly, 'Total_Sales':total,'Week_No.':week_no,'Week_Ending':week_ending}, ignore_index=True)
    
    #converting date to datetime for easy reference
    df['Week_Ending'] = pd.to_datetime(df['Week_Ending'])
    
    return df
```

```python
#testing the function
df1 = perweek(43331)
df1.head(10)
```

```python
#Initializing final dataframe
JapanWeeklydf = pd.DataFrame(columns = ['Name','Platform','Publisher','Genre','Weekly_Sales','Total_Sales','Week_No.','Week_Ending'])
```

```python
#We will now iterate through JapanDateRange to get the full record of Global Sales in vgchartz.com
for page in JapanDateRange:
    JapanWeeklydf = JapanWeeklydf.append(perweek(page), ignore_index=True)
    print('Page index',page,'successfully appended.')
```

```python
JapanWeeklydf.to_csv('./JapanWeeklySales.csv')
```
