```python
#dropping non-residential type zone.
residential = house.loc[house['MSZoning'] != 'C (all)']
residential
```

```python
#MSSubClass is categorical, should convert to string.
residential['MSSubClass'] = residential['MSSubClass'].apply(lambda x : str(x))
residential
```

```python
#Keeping just the non-renovatable features.
residential = residential[["MSSubClass","MSZoning","LotFrontage","LotArea","Street","Alley","LotShape","LandContour","LotConfig","LandSlope","Neighborhood","Condition1","BldgType","HouseStyle","YearBuilt","YearRemodAdd","BsmtQual","1stFlrSF","2ndFlrSF","GrLivArea","BsmtFullBath","BsmtHalfBath","FullBath","HalfBath","BedroomAbvGr","KitchenAbvGr","TotRmsAbvGrd","Fireplaces","GarageType","GarageYrBlt","GarageCars","GarageArea","PoolArea","MiscFeature","MiscVal","YrSold","SalePrice"]]
residential
```

```python
residential.info()
```

```python
#Grouping numerical features into one dataframe 
num_features = residential.loc[:,residential.dtypes != 'object']
num_features.head()
```

```python
#any relationship between numerical values?
fig = plt.subplots(figsize=(17,12))
mask = np.zeros_like(num_features.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(num_features.corr(), mask=mask, annot=True, cmap='RdBu_r', center=0)
```

```python
#Looks like mostly low correlation except for the following pairs:
#GarageArea and GarageCars. Makes sense. We can probably drop one of them.
#GarageYrBlt and YearBuilt. Makes sense. We can probably drop one of them.
num_features = num_features.drop(['GarageCars','GarageYrBlt'], axis=1)
num_features.head()
#TotRmsAbvGrd and GrLivArea. Makes sense. However no. of rooms may still be important
```

```python
#grouping categorical features into another dataframe
cat_feat = residential.loc[:,residential.dtypes == 'object']
#cat_feat['MSSubClass'].value_counts()
cat_feat.columns
```

```python
#Getting dummy variables for categorical features
cat_dummies = pd.get_dummies(cat_feat)
#removing reference categories for each feature
cat_dummies.drop(['MSSubClass_20', 'MSZoning_RH', 'Street_Pave', 'Alley_Pave', 'LotShape_Reg', 'LandContour_Lvl',
       'LotConfig_Inside', 'LandSlope_Gtl', 'Neighborhood_NAmes', 'Condition1_Norm',
       'BldgType_1Fam', 'HouseStyle_1Story', 'BsmtQual_TA', 'GarageType_Attchd', 'MiscFeature_Shed'], axis=1, inplace=True)
cat_dummies.columns
```

```python
#combining the 2 dataframes back into 1 dataframe for further EDA.
combined = num_features.join(cat_dummies)
combined.head()
```

```python
combined[combined.columns[combined.isnull().any() == True]].isnull().sum()
#Null values still present in 'LotFrontage', to be filled in based on train data below
```

```python
#splitting into pre-2010 data and 2010 data, defining train-test matrix
train = combined.loc[combined['YrSold']<2010]
test = combined.loc[combined['YrSold']>=2010]
y_train = train['SalePrice']
X_train = train.drop(['SalePrice'],axis=1)
y_test = test['SalePrice']
X_test = test.drop(['SalePrice'],axis=1)
```

```python
#Filling null values for 'LotFrontage' based on train data only
print(X_train['LotFrontage'].mean())
print(X_train['LotFrontage'].median())
```

```python
#Mean is close to median. Median value will used to fill null values for both test and train datasets.
X_train['LotFrontage'].fillna(X_train['LotFrontage'].median(), inplace=True)
X_test['LotFrontage'].fillna(X_train['LotFrontage'].median(), inplace=True)
```

```python
X_train.head()
```

## Feature Selection  

```python
#using f_classif and chi2
from sklearn.feature_selection import SelectKBest, chi2, f_classif

skb_f = SelectKBest(f_classif, k=5)
skb_chi2 = SelectKBest(chi2, k=5)
#Fit to train data.
skb_f.fit(X_train, y_train)
skb_chi2.fit(X_train, y_train)
#Examine the results.
kbest = pd.DataFrame([X_train.columns, list(skb_f.scores_), list(skb_chi2.scores_)], 
                     index=['feature','f_classif','chi2 score']).T.sort_values('f_classif', ascending=False)
kbest
```

```python
train['SalePrice'][1003]
```

```python
plt.scatter([np.log(x) for x in kbest['f_classif']],
            [np.log(x) for x in kbest['chi2 score']])
```

```python
#Feature selection by recursive features elimination
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
selector = RFECV(lr, step=1, cv=10)
selector = selector.fit(X_train, y_train)
rfecv_columns = np.array(X_train.columns)[selector.support_]
rfecv_columns
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegressionCV

ss = StandardScaler()
Xs_train = ss.fit_transform(X_train)
```

```python
lrcv = LogisticRegressionCV(penalty='l1', Cs=100, cv=10, solver='liblinear')
lrcv.fit(Xs_train, y_train)

coeffs = pd.DataFrame(lrcv.coef_, columns=X_train.columns)
coeffs_t = coeffs.transpose()
coeffs_t.columns = ['lasso_coefs']
coeffs_abs = coeffs_t.abs().sort_values('lasso_coefs', ascending=False)
coeffs_abs
```
